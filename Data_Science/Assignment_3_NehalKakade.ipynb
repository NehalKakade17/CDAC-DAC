```python
#Basic Statistics
#Descriptive Analysis
import pandas as pd
sales_data = pd.read_csv("sales_data_with_discounts.csv")
final_data = pd.DataFrame({"Mean" : sales_data.mean(numeric_only = True),
"Median" : sales_data.median(numeric_only = True),
"Mode" : sales_data.mode(numeric_only = True).iloc[0],
"Standard Deviation" : sales_data.std(numeric_only = True)})
print(final_data)
```

```python
"""Analysis
1) For Volume column of sales data we can say that the people are buying items in lower quantities.Like items in quantities of 3 to 5 are bought.
2) For Average Price column of sales data we can say that it is right skewed. That means the mean is much larger than median. It means the prices of
products are in range that have much wider gap. Like the most comman average price is 400 and least is around 50000. Moreover the standard deviation is
also much more than mean that means a lot of vaiations exsists in this column.
3) For Total Sales Value column of sales data we can say that it is also right skewed. The sales have much wider gap. The frequent sales lies around 24300.
The standard deviation is also more than mean hence a lot of vaiations exists.It means the sales depend on various other factors and and varies a lot.
4) For Discount Rate column of sales data we can say that it is evenly spreaded. Most of the discount given is 5%. And averagely it is between 15-17%.
Moreover the standard deviation is less than mean so the data does not have much variations.
5) For Discount Amount column of sales data we can say that the discount amount is less on frequent items. The standard deviation is also large than mean
that means few users get a large discount amount when they buy more items while those who buy average items get less discount amount. It is around 69.
6) For Net Sales Value column of sales data we can say that it is right skewed. The mean is much more. The standard deviation is also more. The data is
widely spreaded. It means the sales value is less when frequent items are bought and exceeds when costly and more items are bought.
"""
```

```python
import matplotlib.pyplot as plt                                
import pandas as pd
sales_data = pd.read_csv("sales_data_with_discounts.csv")
plt.hist(sales_data['Volume'],bins=30,edgecolor="black")
plt.title("Volume Histogram")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```

```python
# In above Histogram whatever analysis we derived from central tendency measures is visually decribed.
# It shows the highest bar for 3 volume going beyound 100 frequency. While other volume values are less.
```

```python
plt.hist(sales_data['Avg Price'],bins=30,edgecolor="black")
plt.title("Avg Price Histogram")
plt.xlabel("Avg Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# In above Histogram whatever analysis we derived from central tendency measures is visually decribed.
# It shows the highest average price lies between zero and 10000. Also we can see that the data is widely spread making it prone to outliers.
#It is right skewed.
```

```python
plt.hist(sales_data['Total Sales Value'],bins=30,edgecolor="black")
plt.title("Total Sales Value Histogram")
plt.xlabel("Total Sales Value")
plt.ylabel("Frequency")
plt.show()
```
